Model

```python
from sklearn.model_selection import train_test_split

predictors = df.drop("target",axis=1)
target = df["target"]

xtr,xte,ytr,yte = train_test_split(predictors,target,test_size=0.20,random_state=42)
```

```python
print(xtr.shape)
print(xte.shape)
```

```python
print(ytr.shape)
print(yte.shape)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

Random Forest

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
params_rf = {"n_estimators": [50, 100, 200]}
rf_gs = GridSearchCV(rf, params_rf, cv=5)
rf_gs.fit(xtr, ytr)
```

```python
rf_best = rf_gs.best_estimator_
print(rf_gs.best_params_)
```

```python
print('rf: {}'.format(rf_best.score(xte, yte)))
```

```python
rf_pred = rf_gs.predict(xte)
```

```python
plt.figure(figsize=(5, 7))


ax = sns.distplot(y, hist=False, color="r", label="Actual Value")
sns.distplot(rf_pred, hist=False, color="b", label="Fitted Values" , ax=ax)


plt.title('Actual vs Fitted Values for Price')


plt.show()
plt.close()
```

SVM

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear')
svr.fit(xtr, ytr)
```

```python
svr_pred = svr.predict(xte)
```

```python
comp = pd.DataFrame(yte)
```

```python
comp.columns = ['Actual Value']
comp['Predicted Value'] = svr_pred
comp
```

```python
plt.figure(figsize=(5, 7))

ax2 = sns.distplot(yte, hist=False, color="r", label="Actual Value")
sns.distplot(svr_pred, hist=False, color="b", label="Fitted Values" , ax=ax2)

plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(yte, svr_pred))
print('Mean Squared Error:', metrics.mean_squared_error(yte, svr_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(yte, svr_pred)))
```

```python
# Calculate the absolute errors
errors = abs(svr_pred - yte)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')

# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / yte)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

XGBoost

```python
import xgboost as xg
```

```python
xgb = xg.XGBRegressor(objective ='reg:linear', n_estimators = 10, seed = 42)
xgb.fit(xtr,ytr)
```

```python
x_pred = xgb.predict(xte)
```

```python
from sklearn.metrics import mean_squared_error as MSE
rmse = np.sqrt(MSE(yte, x_pred))
print("RMSE : % f" %(rmse))
```

```python
# evaluate xgboost ensemble for regression
from numpy import mean
from numpy import std
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from xgboost import XGBRegressor
# define the model
model = XGBRegressor()
# evaluate the model
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, xtr, ytr, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
# report performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```
